#**Exploratory Data Analysis**

Here, we will import the iris dataset and analyze different aspects using pandas functions and visualization tools.

```python
# Installing required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading dataset
df = sns.load_dataset("iris")
```

```python
#printing first five data
df.head()
```

Printing first five rows of data is usually the first step in EDA.<br> It helps to make sure that our dataset has been loaded properly.

```python
#understanding our data
df.info()
```

.info() helps us to understand our data better.<br>
It gives us basic information about our data like total entries and the columns in our dataset, their types, null values.

```python
#descriptive statistics
df.describe()
```

.describe() gives us an overview of various descriptive statistical measures of the numerical columns of our dataset.<br>
By observing min, max, count, mean, median, standard deviation and so on, <br>
a rough image of our dataset and its distribution is formed on our mind.

```python
#Check missing values
df.isnull().sum()       #there is no missing (null) values in our dataset.
```

```python
#check duplicate values
df.duplicated().sum()       #there is one duplicate row in our dataset
```

```python
#viewing duplicate row
df[df.duplicated()]
```

```python
df.count()
```

```python
#Finding Categorical column value count
df["species"].value_counts()
```

```python
#Finding numeric column value count
df["sepal_length"].value_counts().head()    #since there are many numeric values, we'll be printing only top 5 of them.
```

```python
# Visualize categorical count
plt.figure(figsize=(5, 5))
sns.countplot(x="species",data=df)
plt.title("Count of Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
# Visualize categorical count
plt.figure(figsize=(8, 3))
sns.countplot(x="sepal_length",data=df)
plt.title("Count of Species")
plt.xlabel("Sepal length")
plt.ylabel("Count")
plt.show()
```

```python
#Pair-Plot
sns.pairplot(df,hue="species")
```

Pair-plot is used to visualize the relationship between all pairs of variables.<br> We can see that species are differently separated based on different features.

```python
# Scatter-Plot
sns.scatterplot(x="sepal_length",y="sepal_width",data=df,hue="species")
plt.show()
sns.scatterplot(x="petal_length",y="petal_width",data=df,hue="species")
plt.show()
```

**Observation:**

---
-> It is difficult to distinguish between the species based on sepal_width and sepal_length<br>
-> Iris Setosa can be easily identified due to its smaller petals.<br>
->Iris Virginica and Iris Versicolor slightly overlap, where Virginicia usually having longer and wider petals than Versicolor.

  ---

```python
# Correlation Mattrix
corr=df.corr(numeric_only=True)  #since we have non-numeric column also.
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation between columns")
plt.show()
```

**Observation:**

---
sepal_length and sepal_width : correlation = -0.12
  . It means that when sepal_length increases, sepal_width decreases slightly.
  . Since the value is closer to 0 , it indicates a weak relationship.

petal_length and petal_width : correlation = 0.96
  . It means that when petal_length increases, petal_width also increases.
  . Since the value is closer to 1 , it indicates a strong relationship.

  ---

##**3. Are there any missing values in the dataset?**

```python
int(df.isnull().sum().sum()) > 0
```

###**4. What is the mean, median, and standard deviation of the 'sepal_length' column?**

```python
mean = df['sepal_length'].mean()
median = df['sepal_length'].median()
std = df['sepal_length'].std()

#Printing the values
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)
```

##**5. Which species has the longest average 'petal_length'?**

```python
df[df['petal_length'] == df['petal_length'].max()]['species']
```

###**6. Create a histogram of 'sepal_width'. What does the distribution look like (e.g. normal, skewed)?**

```python
sns.histplot(df['sepal_width'])
```

```python
df['sepal_width'].skew()
```

**Observation**

---
The distribution is slightly right-skewed.<br>
As seen in the histogram, the peak of the curve is not exactly on the center.<br>
On left-side, there are 5 blocks whereas on the right_side, there are 7 blocks.<br>
Since the tail on the right is longer, the distribution is right-skewed or positively skewed.

---

###**7. Use a boxplot to compare the 'petal_length' across different species. What do you observe?**

###**9. What is the correlation between 'sepal_length' and 'petal_length'? Interpret the result.**
