```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
df = pd.read_csv('insurance.csv')
```

```python
#Printing first five rows to ensure that the dataframe has been loaded properly.
df.head()
```

```python
#Understanding the dataset
df.info()
```

```python
#Finding null values in the dataset.

df.isnull().sum().sum()
```

```python
#Duplicate values in the dataset

df.duplicated().sum()
```

```python
#box-plot of age

sns.boxplot(df['age'])
```

```python
df['region'].unique()
```

```python
#Encoding categorical values

df_copy = df.copy()
df_copy['sex'] = df_copy['sex'].map({'female':0,'male':1})
df_copy['smoker'] = df_copy['smoker'].map({'no':0,'yes':1})
df_copy['region'] = df_copy['region'].map({'southwest':0,'southeast':1,'northwest':2,'northeast':3})
df_copy.head()
```

```python
sns.pairplot(df_copy,hue='sex')
plt.show()
```

```python
df_copy['charges'].describe()
```

```python
corr = df_copy[['age','charges','bmi']].corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.regplot(x=df_copy['age'],y=df_copy['charges'])
plt.show()
```

```python
sns.boxplot(x=df_copy['smoker'],y=df_copy['charges'])
plt.title('Smoker vs charges')
plt.show()
```

```python
sns.boxplot(x=df_copy['sex'],y=df_copy['charges'])
plt.title('sex vs charges')
plt.show()
```
